```python
df = pd.DataFrame(license)
df_renamed = df.rename(columns={
    'Applicant ID': 'applicant_id',
    'Gender' : 'gender',
    'Age Group' : 'age_group',
    'Race' : 'race',
    'Training' : 'training',
    'Signals' : 'signals',
    'Yield' : 'yield',
    'Speed Control': 'speed_control',
    'Night Drive' : 'night_drive',
    'Road Signs' : 'road_signs',
    'Steer Control' : 'steer_control',
    'Mirror Usage' : 'mirror_usage',
    'Confidence' : 'confidence',
    'Parking' : 'parking',
    'Theory Test' : 'theory_test',
    'Reactions' : 'reactions',
    'Qualified' : 'qualified'
    })
print(df_renamed.head())
```

```python
df = df_renamed
```

```python
print(df.gender.value_counts(),'\n',
df.age_group.value_counts(),'\n',
df.race.value_counts(),'\n',
df.training.value_counts(),'\n',
df.reactions.value_counts())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
print(df.isnull().sum())
print(df.head())
print(df.shape)
```
